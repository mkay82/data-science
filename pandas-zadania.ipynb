# Zasanie 2
Osób o jakim nazwisku jest najwięcej ?

# Zadanie 3
Po kolumnie z nazwiskiem dodaj kolumnę reprezentującą płeć danej osoby. Można wykorzystać fakt, że imiona
kobiet kończą się literka 'a'.

```python
df["plec"] = df["imie"].apply(lambda x: "kobieta" if x[-1]=='a' else 'mężczyzna')
plec = df.pop('plec')
df.insert(loc=df.columns.get_loc('nazwisko') + 1, column='plec', value=plec)

df.head()
```

# Zadanie 4
Posortuj zbiór danych alfabetycznie po nazwiskach, a w ramach jednego nazwiska po roku urodzenia.

```python
df.sort_values(["nazwisko","rok_ur"], inplace=True)
df.filter(["nazwisko","rok_ur"]).head(10)
```

# Zadanie 5
Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób
słowny, przyjmując:

- Junior (do 3 lat doświadczenia)
- Mid (4 do 8 lat)
- Senior (9 do 12 lat)
- Ekspert (od 13 lat)

```python
def doswiadczenie(lata):
    if lata<=3:
        return "Junior"
    if lata<=8:
        return "Mid"
    if lata<=12:
        return "Senior"
    return "Ekspert"

df["stopien_dosw"] = df.apply(lambda x: doswiadczenie(x["doswiadczenie"]), axis=1)
df.groupby("stopien_dosw").size()
```

# Zadanie 6
Kolumna reprezentująca rok dołączenia do firmy ma puste wartości. Uzupełnij brakujące wartości używając
średniej dla osób o takim samym poziomie doświadczenia (kolumna z Zadania 3), w skali całej firmy.

```python
srednie=round(df.groupby("stopien_dosw")["rok_rozp_pracy"].mean())

print(srednie)
df["rok_rozp_pracy"] = df["rok_rozp_pracy"].fillna(df["stopien_dosw"].map(srednie)).astype(int)

print(df["rok_rozp_pracy"].isna().any())

df
```

# Zadanie 8
Szef chciałby zmienić nazewnictwo departamentów w całej firmie. Napisz kod, który zamieni wartości w
kolumnie reprezentującej departament, nastepująco:

- Software Development -> Software Development Team
- HR -> Human Resources
- Data Science -> Data Analytics

```python
def nowa_nazwa(nazwa):
    if nazwa=="Software Development":
        return "Software Development Team"
    if nazwa == "HR":
        return "Human Resources"
    if nazwa == "Data Science":
        return "Data Analytics"
    return nazwa

df["departament"]=df["departament"].apply(lambda x: nowa_nazwa(x))
df.groupby("departament").size()
```

# Zadanie 10
Które biuro ma najwięcej kobiet na Juniorskim poziomie doświadczenia ?

**UWAGA**: w zbiorze nie ma żadnych juniorów, zmieniłęm warunek na Senior, żeby coś znalazło

```python
junior_k = df[  (df["stopien_dosw"] == "Junior")
                & (df["plec"] == "kobieta") ]
senior_k = df[  (df["stopien_dosw"] == "Senior")
                & (df["plec"] == "kobieta") ]


if len(junior_k.groupby("miasto"))>0:
    print(junior_k.groupby("miasto").size().idxmax())
else:
    print("Wśród zatrudnionych brak kobiet - juniorów")


if len(senior_k.groupby("miasto"))>0:
    print(senior_k.groupby("miasto").size().idxmax())
else:
    print("Wśród zatrudnionych brak kobiet - seniorów")
```

# Zadanie 12
Ile osób, które posiadają obecnie stanowisko seniorskie, dołączyło do firmy w przeciągu ostatnich 10 lat?


```python
from datetime import date
from dateutil.relativedelta import relativedelta

df["data_dolaczenia"]=df.apply(lambda x: date(x["rok_rozp_pracy"], x["miesiac_rozp_pracy"], x["dzien_rozp_pracy"]), axis=1)

dzis = date.today()

senior10 = df[ (df["data_dolaczenia"] < dzis - relativedelta(years=10))
                &(df["stopien_dosw"] == "Senior") ]

print(len(senior10))
senior10
```
